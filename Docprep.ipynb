```python
s = input("Enter any string: ")
empty = ""

for i in s:
    empty = i + empty
print(empty)

if s == empty:
  print("True,Its a Palindrom")
else:
  print("False,not a Palindrom")
```

```python
#count vowels

n = input("Enter any string: ")
vowels = ['a','e','i,o','u']
count = 0

for i in n:
  if i in vowels:
    count += 1
print(count)
```

```python
#remove vowels

n = input("Enter any string: ")
vowels = ['a','e','i','o','u']
without = ""

for i in n:
  if i not in vowels:
    without = without + i
print(without)
```

```python
word_1 = input("Enter first word: ")
word_2 = input("Enter second word: ")

sorted_word1 = sorted(word_1)
sorted_word2 = sorted(word_2)

if sorted_word1 == sorted_word2:
  print("Anagram")
else:
  print("Not Anagram")
```

```python
def array_to_object(arr):
  result = {}
  for key,value in arr:
    result[key]=value
  return result

print(array_to_object([["saketh","rebel"],["abhiram","kohli"]]))
```
